```python
Q1 = df_copy['Electricity usage (kWh)'].quantile(0.25)
Q3 = df_copy['Electricity usage (kWh)'].quantile(0.75)

# Calculate the IQR (Interquartile Range)
IQR = Q3 - Q1

# Define the threshold for identifying outliers
outlier_threshold = 1.5  # Adjust as needed

# Identify outliers using the IQR method
outliers = df_copy[(df_copy['Electricity usage (kWh)'] < Q1 - outlier_threshold * IQR) | (df_copy['Electricity usage (kWh)'] > Q3 + outlier_threshold * IQR)]

# Drop outliers from the DataFrame
df_cleaned = df_copy.drop(outliers.index)
```

```python
cap = 12.5
battery = 0
battery_charge_levels = []

# Loop through each row of the dataframe
for index, row in df_hourly.iterrows():
    # Calculate the net electricity (excess - needed)
    net_electricity = row["Excess solar electricity (kWh)"] - row["Electricity needed (kWh)"]
    # Update the battery charge level by adding the net electricity
    battery += net_electricity
    # If the battery charge level is negative, set it to zero
    if battery < 0:
        battery = 0
    # If the battery charge level is above the cap, set it to the cap
    elif battery > cap:
        battery = cap
    # Append the battery charge level to the list
    battery_charge_levels.append(battery)

# Add the battery charge levels as a new column to the dataframe
df_hourly["Battery charge level (kWh)"] = battery_charge_levels

# Print the dataframe
df_hourly
```

```python
# Scenario 1: Electricity prices increase as expected by the government, 4% p.a.
electricity_needed = [444.969181, 139.190210, 36.446068, 39.449276, 189.243469, 500.295995]
total_electricity_needed = sum(electricity_needed)
cost_of_electricity = [0.17 * total_electricity_needed]
for i in range(20):
    cost_of_electricity.append(cost_of_electricity[-1] * 1.04)
annual_savings_1 = [cost_of_electricity[0]]
for i in range(1, len(cost_of_electricity)):
    annual_savings_1.append(annual_savings_1[-1] + cost_of_electricity[i])

# Scenario 2: Electricity price increases start at 4% p.a. and rise each year by an additional 0.25% p.a., as estimated by Naomi
inflation_rate = 0.04
cost_of_electricity = [0.17 * total_electricity_needed]
for i in range(20):
    cost_of_electricity.append(cost_of_electricity[-1] * (1 + inflation_rate))
    inflation_rate += 0.0025
annual_savings_2 = [cost_of_electricity[0]]
for i in range(1, len(cost_of_electricity)):
    annual_savings_2.append(annual_savings_2[-1] + cost_of_electricity[i])

# Calculate the Net Present Value (NPV) of the future annual savings manually
discount_rate = 0.06 # This value is provided in the question
npv_1 = sum([annual_savings_1[t] / ((1 + discount_rate) ** t) for t in range(len(annual_savings_1))])
npv_2 = sum([annual_savings_2[t] / ((1 + discount_rate) ** t) for t in range(len(annual_savings_2))])

print("Scenario 1: NPV =", npv_1)
print("Scenario 2: NPV =", npv_2)
```

```python
from scipy.optimize import fsolve

initial_cost = 7000

def npv(discount_rate, annual_savings):
    return sum([annual_savings[t] / ((1 + discount_rate) ** t) for t in range(len(annual_savings))])

def irr(initial_cost, annual_savings):
    return fsolve(lambda r: npv(r, annual_savings) - initial_cost, 0.1)[0]

irr_1 = irr(initial_cost, annual_savings_1)
irr_2 = irr(initial_cost, annual_savings_2)

print("Scenario 1: IRR =", irr_1)
print("Scenario 2: IRR =", irr_2)
```
